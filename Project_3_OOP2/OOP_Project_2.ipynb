```python
import os
class Book:
    def __init__(self, title, author, genre, year):
        self.title = title
        self.author = author
        self.genre = genre
        self.year = year

    def __str__(self):
        return f"'{self.title}' by {self.author} ({self.year}) - [{self.genre}]"

class Library:
    def __init__(self):
        self.books = []
        self.filename = "library.txt"
        self.load_books()

    def add_book(self, book):
        self.books.append(book)
        print(f"--> Added successfully: {book.title}")
        self.save_books()

    def remove_book(self, title):
        found = False

        for book in self.books:
            if book.title.lower() == title.lower():
                self.books.remove(book)
                found = True
                print(f"--> Removed successfully: {book.title}")
                self.save_books()
                break
        
        if not found:
            print("--> Error: Book not found!")

    def search_book(self, keyword):
        print(f"\n--- Search Results for '{keyword}' ---")
        found = False
        for book in self.books:
            if keyword.lower() in book.title.lower() or keyword.lower() in book.author.lower():
                print(book)
                found = True
        
        if not found:
            print("No books found.")

    def display_all(self):
        print("\n--- Library Collection ---")
        if not self.books:
            print("Library is empty.")
        else:
            for book in self.books:
                print(book)

    def save_books(self):
        try:
            with open(self.filename, "w") as f:
                for book in self.books:
                    line = f"{book.title},{book.author},{book.genre},{book.year}\n"
                    f.write(line)
        except Exception as e:
            print(f"Error saving file: {e}")

    def load_books(self):
        try:
            if os.path.exists(self.filename):
                with open(self.filename, "r") as f:
                    lines = f.readlines()
                    for line in lines:
                        data = line.strip().split(",")
                        if len(data) == 4:
                            new_book = Book(data[0], data[1], data[2], data[3])
                            self.books.append(new_book)
        except FileNotFoundError:
            print("No saved library found, starting new.")
        except Exception as e:
            print(f"Error loading file: {e}")

def main():
    my_library = Library()

    while True:
        print("\n=== Personal Library Menu ===")
        print("1. Add Book")
        print("2. Remove Book")
        print("3. Search Book")
        print("4. Display All Books")
        print("5. Exit")

        choice = input("Enter choice (1-5): ")

        if choice == '1':
            title = input("Enter Title: ")
            author = input("Enter Author: ")
            genre = input("Enter Genre: ")
            year = input("Enter Year: ")
            new_book = Book(title, author, genre, year)
            my_library.add_book(new_book)

        elif choice == '2':
            title = input("Enter book title to remove: ")
            my_library.remove_book(title)

        elif choice == '3':
            keyword = input("Enter title or author to search: ")
            my_library.search_book(keyword)

        elif choice == '4':
            my_library.display_all()

        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")



if __name__ == "__main__":
    main()
```
