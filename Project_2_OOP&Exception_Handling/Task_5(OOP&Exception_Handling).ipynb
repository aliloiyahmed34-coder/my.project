### Q2.
Create a class named BankAccount:

The balance property should be private.

Deposit method

Withdrawal method

Balance display method

```python
class BankAccount:
    def __init__(self,property):
        self.__property = property
    def Deposit(self,Deposit):
        self.__property += Deposit
    def Withdrawal(self,Withdrawal):
        self.__property -= Withdrawal
    def Balance_display(self):
        print(self.__property)
c1 = BankAccount(20000)
c1.Balance_display()
c1.Deposit(1000)
c1.Balance_display()
c1.Withdrawal(12000)
c1.Balance_display()
```

```python
class BankAccount:#Another solution
    def __init__(self, name, initial_balance):
        self.name = name
        self.__balance = initial_balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print('An amount of:',amount)
        else:
            print('Incorrect amount')
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print('An amount was withdrawn:',amount)
        else:
            print('Sorry, I don\'t have enough credit.')
    def display_balance(self):
        print('Account holder:',self.name)
        print('Current balance:',self.__balance)
my_account = BankAccount("Ahmed", 1000)
my_account.deposit(500)
my_account.withdraw(200)
my_account.display_balance()
```

### Q3.
Create a Class Person containing a name, and a Class Employee that inherits from it and adds salary.

```python
class Person:
    def name1(self, name):
        self.name = name

class Employee(Person):
    def salary1(self, salary):
        self.salary = salary
    def print1(self):
        print('Employee name:',self.name,'Employee salary:',self.salary)

e1 = Employee()
e1.name1('Ahmed')
e1.salary1(20000)
e1.print1()
```

```python
class Person:#Another solution
    def __init__(self, name):
        self.name = name
    def display_info(self):
        print('name:',self.name)
class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary
    def display_salary(self):
        print('salary:' ,self.salary)
emp1 = Employee('Ahmed', 20000)
emp1.display_info()
emp1.display_salary()
```

### Q4.
Write a function that accepts an object and calls a method named speak

```python
class Cat:
    def speak(self):
        print("Meow! Meow!")
class Dog:
    def speak(self):
        print("Woof! Woof!")
def creature_speak(creature):
    creature.speak()
cat1 = Cat()
dog1 = Dog()
print("Cat")
creature_speak(cat1)
print("Dog")
creature_speak(dog1)
```

## Exception_Handling
### Q1.
Write a program that divides two numbers and catches the division by zero error.

```python
try:
    a = int(input("please enter a number"))
    b = int(input("please enter a number"))
    print(a/b)
except ZeroDivisionError as h:
    print("An error occured.")
    print(h)

print("Hello World")
```

### Q2.
Handle:

ValueError

ZeroDivisionError

```python
try:
    a = int(input("please enter a number"))
    b = int(input("please enter a number"))
    print(a/b)
except ZeroDivisionError as h1:
    print("An error occured.")
    print(h1)
except ValueError as h2:
    print("An error occured.")
    print(h2)
print("Hello World")
```
